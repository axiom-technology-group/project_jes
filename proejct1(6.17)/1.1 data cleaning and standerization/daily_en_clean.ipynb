## Overview of datacleaning:
### 0.0 read the craped csv file
### 1.0 drop all the null values
### 1.1 clean the source column
### 1.2 combine all the needed article
### 1.3 grab all the meanful content( length above 30)

```python
import pandas as pd
import numpy as np
```

## 0.0 read the craped csv file

```python
df = pd.read_csv("dailyen.csv")
```

```python
df.head()
```

```python
# describe the current dataset
df.describe()
```

```python
#drop all the missing value
df_en = df.dropna()
```

```python
df_en.describe()
```

```python
df_en.head()
```

```python
df_en.head(10)
```

```python
#split the "| \n" from source column
df_en['source'] = df_en['source'].map(lambda x: x.split("| \n"))                  
```

```python
df_en['source'].head()
```

```python
def func_source(x):
    if x[0].strip()== x[-1].strip():
        return x[0]
    else:
        return x[0].strip()+"," + x[-1].strip()
df_en['source'].apply(func_source)
df_en['source'] = df_en['source'].apply(func_source)
```

```python
df_en.head()
```

```python
# replace all the "by" in this column and let all the characters uppercase
df_en['source'] = df_en['source'].map(lambda x: x.replace("By","").upper()) 
```

```python
df_en.head()
# sor the values by using the url, so all rows that are about the same article will be in order.
df_en = df_en.sort_values(by=['url'])
```

```python
# join all the same articles from different pages
df_en['content'] = df_en.groupby(['publishtime','title'])['content'].transform(lambda x: ','.join(x))
```

```python
df_en.describe()
#drop duplicates based on the first column and keep the first one
df_en = df_en.drop_duplicates('content', keep='first')
```

```python

df_en.head()
```

```python
# then reorder the index by colum_1 and column_2.
df_en = df_en.sort_values(by=['column_1','column_2'])
df_en = df_en.reset_index()
df_en.head()
```

```python
df_en['content'] = df_en['content'].map(lambda x: x.replace("\xa0","")) 
df_en['content'] = df_en['content'].map(lambda x: x.replace(" ,","")) 
df_en['content'] = df_en['content'].map(lambda x: x.lstrip(':')) 
df_en.head()
```

```python
# get the corrent content which let the content length above 10
def content_source(x):
    if len(x.strip()) <= 30:
        return np.nan
    else:
        return x
df_en['content'] = df_en['content'].apply(content_source)
```

```python
# then drop all the null value
df_en = df_en.dropna()
del df_en['index']
df_en.head()
```

```python
#finally export the data and put in csv file.
df_en.to_csv('ChinaDaily_EN_cleaned.csv')
```
