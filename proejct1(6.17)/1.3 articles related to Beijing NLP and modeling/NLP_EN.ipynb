```python
def top_freq(string,num = 20):
    doc = nlp(string)
# all tokens that arent stop words or punctuations
    words = [token.text for token in doc if token.is_stop != True and token.is_punct != True]

# noun tokens that arent stop words or punctuations
    nouns = [token.text for token in doc if token.is_stop != True and token.is_punct != True and token.pos_ == "NOUN"]
    
    return word_count(nouns,num)
```

```python
df_en['top_nouns'] = df_en['content'].apply(top_freq)
```

```python
def havebeijing(lst):
    if 'Beijing' in lst:
        return True
    else:
        return False
    
def havelifestyle(lst):
    if 'travel' in lst.lower() or 'food' in lst.lower() or 'culture' in lst.lower() or 'sport' in lst.lower():
        return True
    else:
        return False
    
def haveeducation(lst):
    if 'school' in lst.lower() or 'class' in lst.lower() or 'study' in lst.lower() or 'teach' in lst.lower():
        return True
    else:
        return False

def havetech(lst):
    if 'internet' in lst.lower() or 'ai' in lst.lower() or 'tech' in lst.lower() or 'online' in lst.lower():
        return True
    else:
        return False

        
```
