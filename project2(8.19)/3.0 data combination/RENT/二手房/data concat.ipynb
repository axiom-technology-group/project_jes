```python
import pandas as pd
import numpy as np

df1 = pd.read_csv('上海租房地址数据.csv')
df2 = pd.read_csv('北京租房地址数据.csv')
df3 = pd.read_csv('深圳租房地址数据.csv')
df4 = pd.read_csv('成都租房地址数据.csv')
df5 = pd.read_csv('广州租房地址数据.csv')

df1['city'] = '上海'
df2['city'] = '北京'
df3['city'] = '深圳'
df4['city'] = '成都'
df5['city'] = '广州'
```

```python
df = pd.concat([df1,df2,df3,df4,df5])
```
