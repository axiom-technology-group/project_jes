```python
df = pd.read_csv('深圳商铺数据.csv')
```

```python
df.head()
```

```python
def havemi(string):
    if "米" in string and len(string) <= 7:
        return True
    else:
        return False
        

df['pend'] = df['loupandizhi'].apply(havemi)

list_index = df[df['pend'] == True].index.tolist()

df_temp = df[df['pend'] == True]

df_temp.head()


```

```python
list_loupandizhi = df_temp['loupandizhi'].tolist()

df.iloc[list_index,8] = list_loupandizhi

list_loupanmingcheng = df_temp['loupanmingcheng'].tolist()

df.iloc[list_index,6] = list_loupanmingcheng

list_shihejingying = df_temp['shihejingying'].tolist()

df.iloc[list_index,7] = list_shihejingying

df.iloc[list_index,13] = np.nan

list_miankuan = df_temp['miankuan'].tolist()

df.iloc[list_index,4] = list_miankuan

list_jinshen = df_temp['jinshen'].tolist()

df.iloc[list_index,0] = list_jinshen

```

```python
df[df['pend'] == True]
```

```python
def delshangpu(string):
    if '商铺' in string:
        return string.replace('商铺','').strip()
    
    elif '餐饮街铺' in string:
        return string.replace('餐饮街铺','').strip()
    
    elif '?' in string:
        return string.replace('?','').strip()
    
    elif '\r\n' in string:
        return string.replace('\r\n','').strip()
            
    else:
        return string.strip()

df['loupanmingcheng'] = df['loupanmingcheng'].apply(delshangpu)

list_temp = df['loupanmingcheng'].tolist()
list_temp2 = df['location'].tolist()
list_location = []
for i in range(len(list_temp)):
    list_location.append(str(list_temp2[i])+','+str(list_temp[i]))
df['location'] = list_location

del df['pend']
```

```python
df_temp = df.copy()

def haveceng(string):
    if '层' in str(string):
        return str(string)
    else:
        return np.nan
df['floor'] = df_temp['floor'].apply(haveceng)

def havem2(string):
    if '㎡' in str(string):
        return str(string)
    else:
        return np.nan
df['shiyongmianji'] = df_temp['floor'].apply(havem2)
```

```python
def nottolong(string):
    if len(string) > 16:
        return False
    else:
        return True

df = df[df['loupanmingcheng'].apply(nottolong)]
```

```python
df.to_csv('深圳商铺数据.csv',index = False, encoding = 'utf-8-sig')
```
