```python
import pandas as pd
import numpy as np
import requests
```

```python
df = pd.read_csv('成都租房数据.csv')
```

```python
df.head(2)
```

```python
list_location = df['location'].value_counts().index.tolist()
dict_location = {}
Key = '6b95b84cab6100c0a6f52c62ece97e4b'
My_key = 'f654d644aeae0e1da847d7ca408f4370'
url_pre = 'https://restapi.amap.com/v3/geocode/geo?key={}&address={}&city={}'
counter = 0
for i in list_location:
    location = i.split(',')[-1]
    url = url_pre.format(Key, location, '成都')
    res = requests.get(url).json()
    if res['status'] == '0':
        dict_location[i] = np.nan
        counter +=1
    else:
        try: 
            dict_location[i] = res['geocodes'][0]['location']
            if counter % 50 == 0:
                print('one batch(50) processed')
        except:
            dict_location[i] = np.nan
            
        counter +=1
```

```python
df['location_coords'] = np.nan
counter = 0
for i in list_location:
    counter += 1
    df_temp = df[df['location'] == i]
    df_temp_index = df_temp.index.tolist()
    df.iloc[df_temp_index,22] = dict_location[i]
    if counter %50 == 0:
        print('A batch(50) is done')
```

```python
df.to_csv('成都租房地址数据.csv', encoding = 'utf-8-sig',index = False)
```
